```python
respuesta2 = requests.get('https://api.github.com/repos/ironhack-datalabs/scavenger/git/trees/master?recursive=1')
data = respuesta2.json()
data
```

```python
paths = []
for i in data['tree']:
    if i['path'].endswith('scavengerhunt'):
        paths.append(i['path'])
paths.sort(key= lambda x: x.split('/.')[1])
```

```python
paths
```

```python
joke_base64 = []
for i in paths:
    respuesta = requests.get('https://api.github.com/repos/ironhack-datalabs/scavenger/contents'+i, auth=(username,token))
    data = respuesta.json()
    joke_base64.append(data['content'])
```

```python
print(joke_base64)
```

```python
import base64
import re
joke = [base64.b64decode(i.encode()).decode('utf-8') for i in joke_base64]
joke_final = re.sub('\n','',' '.join(joke))
```
