```python
import requests
import base64
busqueda=requests.get("https://api.github.com/repos/ironhack-datalabs/scavenger/git/trees/master?recursive=1")
```

```python
data=busqueda.json()
data
```

```python
data.keys()
```

```python
data["tree"]
```

```python
lista=[]
for i in range(len(data["tree"])):
    file=data["tree"][i]
    if file["path"].endswith(".scavengerhunt"):
        lista.append((file["path"]))
lista
```

```python
lista1=sorted(lista, key=lambda string:int(string[9:11]))
lista1
```

```python
palabras=[]
for path in lista1:
    busqueda=requests.get(f"https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{path}")
    texto=busqueda.json()
    palabras.append(texto)
palabras
```

```python
chiste=[(base64.b64decode(palabra["content"].encode('ascii'))).decode('ascii').replace('\n','')for palabra in palabras]
" ".join(chiste)
```
